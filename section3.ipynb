### The QQ Plot

    known as the quantile-quantile plot. It is used to determine whether a dataset is distributed a certain way

```python
scipy.stats.probplot(df['spx'], plot = pylab)
pylab.show()
```

    Theoretical quantiles = standard disviations
    red line - what the data should follow if they are normally distributed
